```python
import matplotlib.pyplot as plt
```

```python
#MergeSort

merge_2 = []
merge_4 = []

with open("MergeSort_2.txt") as f:
    merge_2 = f.read().splitlines()
    
with open("MergeSort_4.txt") as f:
    merge_4 = f.read().splitlines()
    
merge_1 = []
size = []

for i in range(len(merge_2)):
    size.append(2**(i+1))
    vals = merge_2[i].split(" ")
    merge_1.append(float(vals[0]))
    merge_2[i] = float(vals[1])
    
for i in range(len(merge_4)):
    vals = merge_4[i].split(" ")
    merge_4[i] = float(vals[1])

plt.rcParams['figure.figsize'] = [20, 10]

plt.plot(size, merge_1, alpha=1, c = "green", marker = ',', linewidth =1, label="Merge_1")
plt.plot(size, merge_2, alpha=1, c = "greenyellow", marker = ',', linewidth =1, label="Merge_2")
plt.plot(size, merge_4, alpha=1, c = "gold", marker = ',', linewidth =1, label="Merge_4")

plt.xlabel("Array size")
plt.ylabel("Time")
plt.legend(markerscale=1, frameon=True)
plt.show()   
```
